```python
from matplotlib import pyplot as plt
x1 = [1,6,5]
y1 = [4,2,3]
x2 = [8,1,3]
y2 = [1,2,6]
plt.plot(x1,y1)
plt.plot(x2,y2)
plt.show()
```

```python
from matplotlib import pyplot as plt
from matplotlib import style
style.use('ggplot')
x1 = [1,6,5]
y1 = [4,2,3]
x2 = [8,1,3]
y2 = [1,2,6]
plt.plot(x1,y1, label='first')
plt.plot(x2,y2, label= 'second')
plt.legend()
plt.show()
```

```python
from matplotlib import pyplot as plt
from matplotlib import style
style.use('ggplot')
x1 = [1,6,5]
y1 = [4,2,3]
x2 = [8,1,3]
y2 = [1,2,6]
plt.plot(x1,y1, label='first', linewidth= 4)
plt.plot(x2,y2, label= 'second')
plt.legend()
plt.show()
```

```python
from matplotlib import pyplot as plt
from matplotlib import style
style.use('ggplot')
x1 = [1,6,5]
y1 = [4,2,3]
x2 = [8,1,3]
y2 = [1,2,6]
plt.scatter(x1,y1, label='first', linewidth= 4)
plt.hist(x2,y2, label= 'second')
plt.legend()
plt.show()
```
